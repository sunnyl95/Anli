```python
# 登记时间减两年
import pandas as pd
import datetime
def setDatetime(data_Org):
    # 时间特征：object转datetime
    data_Org['djrq']=pd.to_datetime(data_Org['djrq'])
    # 日期减两年
    data_Org["djrq"] = data_Org["djrq"] - datetime.timedelta(days = 365*2)
    # datetime转object
    str_date = []
    for i in range(len(data_Org['djrq'])):
        str_date.append(data_Org['djrq'][i].strftime("%Y-%m-%d"))
    data_Org['djrq'] = str_date
    # return date_Org
```

```python
# 缺失值 & 异常值 处理
def setErrorValue(df):
    for i in df.columns.values.tolist():
        for j in range(len(df[i])):
            # 缺失值
            if df[i][j] == '':
                df[i][j] = 0
            # 异常值
            if df[i][j] == '""""""""""""""':
                df[i][j] = 0
```

```python
# 删除结果影响较小的列
def delCol(df):
    # 纳税人名称  注册地址  生产经营地址  街道乡镇  经营范围  注册资本  从业人数
    columns = ["nsrmc","zcdz","scjydz","xzjd","jyfw","zczb","cyrs"]
    df.drop(columns,axis=1,inplace=True)
```

```python
# 获取算法需要的数据
def getData(df):
    columns = []
    for i in features_train.columns.values.tolist():
        # 季度
        if "q" in i:
            columns.append(i)
        # 行业
        if "hydl" in i:
            columns.append(i)
        if "hyml" in i:
            columns.append(i)
        # 人员姓名 & 手机号
        if "xm" in i:
            columns.append(i)
    data = df[columns]
    # 对行业大类编码
    columns = ["hydl"]
    data_dummy = pd.get_dummies(data[columns])
    data = data.join(data_dummy)
    # 对行业门类编码
    columns = ["hyml"]
    data_dummy = pd.get_dummies(data[columns])
    data = data.join(data_dummy)
    return data
```
