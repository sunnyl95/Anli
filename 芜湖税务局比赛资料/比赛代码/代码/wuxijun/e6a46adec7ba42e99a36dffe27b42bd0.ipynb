```python
# machine learning
import msgpack
import numpy as np
from sklearn import svm, tree, ensemble
import xgboost
import time


class MLClassification(object):
    def __init__(self, method):
        self.train_data, self.eval_data, self.test_data = self.get_dataset()
        self.threshold = 0.5
        if method == 'svm':
            self.clf = svm.SVC(probability=True)
        elif method == 'tree':
            self.clf = tree.DecisionTreeClassifier()
        elif method == 'forest':
            self.clf = ensemble.RandomForestClassifier()
        elif method == 'xgboost':
            self.clf = xgboost.XGBClassifier()

    def get_dataset(self):
        datasets = []
        for dataset in ('train', 'eval', 'test'):
            with open(f'{dataset}_list.txt', 'rb') as f:
                datasets.append(msgpack.load(f))
        return datasets

    def train(self):
        data = [x[0] for x in self.train_data]
        label = [x[1] for x in self.train_data]
        data = np.array(data)
        label = np.array(label)
        self.clf.fit(data, label)
        predict = self.clf.predict(data)
        return sum(np.array(predict) == np.array(label)) / len(label)

    def eval(self):
        data = [x[0] for x in self.eval_data]
        label = [x[1] for x in self.eval_data]
        data = np.array(data)
        label = np.array(label)
        predict = self.clf.predict(data)
        probility = self.clf.predict_proba(data)
        score = F.cross_entropy(t.from_numpy(probility), t.from_numpy(label), reduction='sum') / label.sum()
        true_positive = sum(predict * label)
        false_positive = sum(predict * (1 - label))
        true_negative = sum((1 - predict) * (1 - label))
        false_negative = sum((1 - predict) * label)
        return f'tp:{true_positive}, fp:{false_positive}, tn:{true_negative}, ' \
               f'fn:{false_negative}, ' \
               f'p:{true_positive * 100 / (true_positive + false_positive):.2f}% ' \
               f'r:{true_positive * 100 / (true_positive + false_negative):.2f}% ' \
               f'score:{(1-score)*100:.2f}'


    def test(self):
        data = [x[0] for x in self.test_data]
        data = np.array(data)
        predict = self.clf.predict_proba(data)
        predict = np.clip(predict, a_min=0, a_max=1)
        with open('ml_result.txt', 'w', encoding='utf-8') as f:
            f.write(','.join([f'{i[1]}' for i in predict.tolist()]))
        return [i[1] for i in predict.tolist()]


if __name__ == '__main__':
    for machine_method in ['forest', 'xgboost', 'tree', 'svm'][3:4]:
        print(f'method: {machine_method}')
        start = time.perf_counter()
        ml_classifier = MLClassification(machine_method)
        ml_classifier.train()
        print(ml_classifier.eval())
        
        get_score.post_user_id('1')
        print(f'time used:{time.perf_counter() - start:.2f}s\n')
        predicted = ml_classifier.test()
```

```python
# deep learning
import argparse
import random
import time

import torch as t
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.optim.lr_scheduler import StepLR

class AverageMeter(object):
    def __init__(self):
        self.mean = 0
        self.count = 0

    def add(self, num, count=1):
        self.mean = (self.mean * self.count + num * count) / (self.count + count)
        self.count += count

    def reset(self):
        self.mean = 0
        self.count = 0

    @property
    def value(self):
        return self.mean


class MLP(nn.Module):
    def __init__(self, input_feature=98):
        super().__init__()
        self.linear_1 = nn.Linear(input_feature, 400)
        nn.init.kaiming_normal_(self.linear_1.weight)
        self.batch_norm_1 = nn.BatchNorm1d(400)
        self.linear_2 = nn.Linear(400, 200)
        nn.init.kaiming_normal_(self.linear_2.weight)
        self.batch_norm_2 = nn.BatchNorm1d(200)
        self.linear_3 = nn.Linear(200, 2)
        nn.init.kaiming_normal_(self.linear_3.weight)

    def forward(self, x):
        x = self.linear_1(x)
        x = self.batch_norm_1(x)
        x = t.relu(x)
        x = self.linear_2(x)
        x = self.batch_norm_2(x)
        x = t.relu(x)
        x = t.sigmoid(self.linear_3(x))
        return x


class EnterpriseDataset(t.utils.data.Dataset):
    def __init__(self, dataset='train'):
        super().__init__()

        with open(f'{dataset}_list.txt', 'rb') as f:
            self.data = msgpack.load(f)
        self.data = [(t.tensor(x, dtype=t.float32),
                      t.tensor(y, dtype=t.long))
                     for x, y in self.data]

    def __getitem__(self, item):
        return self.data[item]

    def __len__(self):
        return len(self.data)


def parse_args():
    parser = argparse.ArgumentParser(description='args for network')
    parser.add_argument('--batch-size', help='batch size', default=10, type=int)
    parser.add_argument('--lr', help='learning rate', default=1e-3, type=float)
    parser.add_argument('--optimizer', help='optimize method for gradient descend',
                        choices=['Adam', 'SGD'], default='SGD')
    parser.add_argument('--cuda', help='whether to use cuda', action='store_true')
    parser.add_argument('--iter', help='max training iterations', default=10000, type=int)
    parser.add_argument('--seed', help='seed for random', default=0, type=int)
    parser.add_argument('--print-step', help='print info per print_step iteration',
                        default=300, type=int)
    parser.add_argument('--save-step', help='save model per save_step iteration',
                        default=10000, type=int)
    parser.add_argument('--checkpoint', help='load model from iteration',
                        default='final', type=str)
    parser.add_argument('--threshold', help='threshold for final result',
                        default=0.5, type=float)
    parser.add_argument('--positive-weight', help='weight for positive labels',
                        default=0.5, type=float)
    parser.add_argument('--eval-only', help='eval without training (default:train and eval)',
                        action='store_true')
    arg = parser.parse_args()

    setattr(arg, 'device', t.device('cuda') if arg.cuda else t.device('cpu'))
    return arg


def ohem(predict, label, arg, top=10):
    loss_weight = t.tensor([1 - arg.positive_weight, arg.positive_weight], device=arg.device)
    loss = F.cross_entropy(predict, label, weight=loss_weight, reduction='none')
    loss = loss.topk(top)[0]
    return loss.mean()


def train(arg):
    train_dataset = EnterpriseDataset('train')
    eval_dataset = EnterpriseDataset('eval')
    dataloader = DataLoader(dataset=train_dataset, batch_size=arg.batch_size,
                            shuffle=True, drop_last=False, pin_memory=True)
    eval_dataloader = DataLoader(dataset=eval_dataset, batch_size=arg.batch_size,
                                 shuffle=False, drop_last=False, pin_memory=True)
    model = MLP(input_feature=len(train_dataset[0][0]))
    model.to(arg.device)
    model.train()
    optimizer = getattr(t.optim, arg.optimizer)(model.parameters(), arg.lr)
    scheduler = StepLR(optimizer, step_size=arg.iter // 3, gamma=0.1)

    iteration = 0
    loss_meter = AverageMeter()
    true_positive = AverageMeter()
    true_negative = AverageMeter()
    false_positive = AverageMeter()
    false_negative = AverageMeter()
    for _ in (range(arg.iter * arg.batch_size // len(train_dataset))):
        for i, (data, label) in enumerate(dataloader):
            optimizer.zero_grad()
            data = data.to(arg.device)
            label = label.to(arg.device)
            predict = model(data)
            loss = ohem(predict, label, arg)
            loss.backward()
            optimizer.step()
            scheduler.step()
            loss_meter.add(loss.cpu().data)
            label = label.detach().cpu()
            predict = predict.detach().cpu().argmax(dim=1)
            true_positive.add(sum(predict * label) * 1.)
            true_negative.add(sum((1 - predict) * (1 - label)) * 1.)
            false_positive.add(sum(predict * (1 - label)) * 1.)
            false_negative.add(sum((1 - predict) * label) * 1.)
            iteration += 1
            if not iteration % arg.print_step:
                train_precision = true_positive.value / (true_positive.value + false_positive.value)
                train_recall = true_positive.value / (true_positive.value + false_negative.value)
                train_f = 2 / (1 / train_precision + 1 / train_recall)

                model.eval()
                with t.no_grad():
                    eval_predict_list = t.zeros(0, 2)
                    eval_label_list = t.zeros(0, dtype=t.long)
                    for j, (eval_data, eval_label) in enumerate(eval_dataloader):
                        eval_data = eval_data.to(arg.device)
                        eval_label_list = t.cat((eval_label_list, eval_label))
                        eval_predict = model(eval_data).cpu()
                        eval_predict_list = t.cat((eval_predict_list, eval_predict))
                eval_loss = 1. - F.cross_entropy(eval_predict_list, eval_label_list,
                                                 reduction='sum') / eval_label_list.sum()
                eval_accu = (eval_predict_list.argmax(dim=1) == eval_label_list).sum() * 100. / len(eval_label_list)

                print(f'iter:{iteration}, loss:{loss_meter.value:.5f}, '
                      f'p:{train_precision * 100.:.2f}% '
                      f'r:{train_recall * 100.:.2f}% '
                      f'f:{train_f * 100.:.2f}% '
                      f'eval_accu:{eval_accu:.2f}% '
                      f'eval_score:{eval_loss * 100:.2f}')
                loss_meter.reset()
                true_positive.reset()
                true_negative.reset()
                false_positive.reset()
                false_negative.reset()
                model.train()
            if not iteration % arg.save_step and iteration:
                t.save(model.state_dict(), f'checkpoint/mlp_{iteration}.pkl')
            t.save(model.state_dict(), 'checkpoint/mlp_final.pkl')


def eval(arg):
    dataset = EnterpriseDataset('eval')
    dataloader = DataLoader(dataset=dataset, batch_size=arg.batch_size,
                            shuffle=False, drop_last=False, pin_memory=True)

    model = MLP(input_feature=len(dataset[0][0]))
    model.load_state_dict(t.load(f'checkpoint/mlp_{arg.checkpoint}.pkl'))
    model.to(arg.device)
    model.eval()
    predict_list = t.zeros(0, 2)
    all_label = []
    start = time.perf_counter()
    with t.no_grad():
        for i, (data, label) in enumerate(dataloader):
            data = data.to(arg.device)
            label = label.to(arg.device)
            predict = model(data)
            predict_list = t.cat((predict_list, predict.cpu()))
            all_label.extend(label.cpu().tolist())
    predict = predict_list.argmax(dim=1)
    label = t.tensor(all_label)
    true_positive = sum(predict * label)
    false_positive = sum(predict * (1 - label))
    true_negative = sum((1 - predict) * (1 - label))
    false_negative = sum((1 - predict) * label)
    score = 1. - F.cross_entropy(predict_list, label, reduction="sum") / label.sum()
    print(f'tp:{true_positive}, fp:{false_positive}, tn:{true_negative}, '
          f'fn:{false_negative},time used:{time.perf_counter() - start:.2f}s, '
          f'device: {arg.device} '
          f'score:{score.data * 100:.2f}')


def test(arg):
    dataset = EnterpriseDataset('test')
    dataloader = DataLoader(dataset=dataset, batch_size=arg.batch_size,
                            shuffle=False, drop_last=False, pin_memory=True)

    model = MLP(input_feature=len(dataset[0][0]))
    model.load_state_dict(t.load(f'checkpoint/mlp_{arg.checkpoint}.pkl'))
    model.to(arg.device)
    model.eval()
    result = []
    with t.no_grad():
        for i, (data, _) in enumerate(dataloader):
            data = data.to(arg.device)
            predict = model(data)
            result.extend(predict[:, 1].data.tolist())
    with open('result.txt', 'w') as f:
        f.write(','.join([f'{x}' for x in result]))


if __name__ == '__main__':
    arg = parse_args()
    random.seed(arg.seed)
    t.manual_seed(arg.seed)

    #print(arg)
    #if not arg.eval_only:
    #    train(arg)
    #eval(arg)
    #test(arg)
```
