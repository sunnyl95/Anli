```python
import pandas as pd
df = pd.read_csv("./origin_dataset/train.csv", encoding="GBK")
```

```python
df.head(5)   # 敏感字段："zcdz", "scjydz", "nsrmc", "xzjd", "bsrxm", "fddbrxm"
```

```python
from hashlib import md5
#对字符串进行md5加密处理
def encrypt_md5(s):
    #转成字符串
    s = str(s)
    
    # 创建md5对象
    new_md5 = md5()
    new_md5.update(s.encode(encoding='utf-8'))
    
    # 加密
    return new_md5.hexdigest()

```

```python
## 对敏感字段进行md5加密
def md5_encrypted_processor(data, feature):
    data[feature] = data.apply(lambda x: encrypt_md5(x[feature]), axis=1)
    return data
```

```python
sensitive_features = ["zcdz", "scjydz", "nsrmc", "xzjd", "bsrxm", "fddbrxm"]
```

```python
for sf in sensitive_features:
    df = md5_encrypted_processor(df, sf)
```

```python
df.head(5)   # 敏感字段："zcdz", "scjydz", "nsrmc", "xzjd"已加密
```

```python
#将加密后的数据保存下来
df.to_csv("./dataset/train.csv", encoding="GBK")
print("加密成功，并保存！")
```
